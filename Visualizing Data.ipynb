## Start with a basic graph
A graph or chart is just a visualization of numeric data. One of the reasons for displaying your data using MatPlotLib is flexibility. You can combine as many visual elements as you need rather than being limited to preset graphic types chosen by someone else (like in Excel, for example).

### Defining the plot
The basis of any graph is the plot. We define the plot with three components:
- data values
- the `matplotlib.pyplot` module
- the type of display you want.


````python
import matplotlib.pyplot as plt
values = [1, 5, 8, 9, 2, 0, 3, 10, 4, 7]
values2 = [3, 8, 9, 2, 1, 2, 4, 7, 6, 6]
````

````python
# Example: basic line graph
plt.plot(range(10,20), values)
plt.show()
````

The previous cell creates a simple line graph using the `plot()` method of the `matlibplot.pyplot` module. The statement 
```
plt.plot(range(1,11), values)
```
says to create a plot using the values 1 through 10 as the x-axis values (remember how `range()` works?) and the elements of the `values` list as the y-axis values. The `plt.show()` statement displays the resulting plot to the screen.

How do we draw multiple lines? Call `plt.plot()` for each line we want.

````python
# Example: multiple lines on one graph
plt.plot(range(10,20), values)
plt.plot(range(10,20), values2)
plt.show()
plt.savefig('MyPlot.png', format = 'png')
````

## Going beyond the basic chart
There are many options available to you for visualizing data. What follows is just an introduction to some of the most often used charts, but you can explore other options at the `matplotlib` documentation link.

It is up to you as the analyst to select an appropriate type of visualization for you data. A good source for a discussion of how to select the best type of graphic and then use it to tell a story about your data is the book **_Storytelling with Data: A Data Visualization Guide for Business Professionals_** by Cole Nussbaumer Knaflic. 



### Choosing the right graph
Often the most important thing you do with a visualization is when you decide which visualization to use. That decision begins with the purpose of your analysis.

Your analysis (and subsequent communication) may be **exploratory** or **explanatory**.
- Exploratory  working through an analysis to see if something is there.
- Explanatory  you found something and you want to tell people about it.

Dont just throw all of your data on a screen and hope the audience reaches the right conclusion. Guide them to it.

---
Six key lessons for good visualizations.
1. Understand the context.
    - Who is your audience?
    - What do you want your audience to know or do?
        - The big idea
        - The 3-minute story or elevator pitch
    - How will you be communicating with your audience? Live talks and written work allow different levels of control.
2. Choose an appropriate visual display.
    - Text taps into the verbal system of the audience. 
    - Pictures use the visual system, which is faster.
        - Charts and graphs arent automatically always better, though.
        - If you have just a number or two you want to emphasize, use them with a little explanatory text to get your point across.
    - Things to avoid:
        - pie charts
        - donut charts
        - 3D
        - using a second y-axis
3. Eliminate clutter.
    - **Clutter** is the visual elements that take up space but dont add anything to the audiences understanding. 
    - We want to distinguish between **signal** and **noise**. Signal is what we want to communicate, noise is what gets in the way of that communication, aka clutter. 
    - Once you have a completed visual, see if you can remove any parts without changing the message. If you can, those parts are clutter.
4. Focus attention where you want it.
    - With a chart or graph, the audience will typically start at the top-left and work down in a Z pattern. 
    - Use that by putting important stuff at the top.
    - To direct the audience's attention, use **_preattentive attributes_**. 
        - They tell the audience where to look. 
        - They enable the audience to see what we want them to see before they even know they are seeing it. 
        - We can utilize differences in: orientation, shape, line length, line width, size, curvature, extra marks or enclosure, color, intensity, or position in space. 
        - It also creates a visual hierarchy, an order for when different items should be seen.
5. Think like a designer.
    - Good design has to be planned. 
    - Keep it simple. 
    - Using text wisely can have a big impact. 
    - Be smart with color and white space.
        - Use color sparingly. The use of many different colors causes the data to get lost. Use differing shades of a single color when possible.
        - Roughly 8% of men (and % of women) are colorblind and usually have a hard time with shades of red and green, so it is best to avoid those colors.
        - Colors also invoke emotions so watch which colors you use. Some colors can have different meanings between cultures.
    - Make sure the design of your visualization will be accepted by your audience.
6. Tell a story.
    - Why? People remember stories.
    - Stories express how and why life changes. 
        - They start with balance, then something upsets the balance. 
        - They have a beginning, a middle, and an end.

## Pie charts: parts of a whole
Pie charts show the percentage or fraction of the whole set of data each part represents. The reason why they are good to use in some situations is the same reason they are generally overused: they are easy to understand.

A simple pie chart can be built using just the values. (In fact, all of the matplotlib chart can be drawn using just data and the default parameter values).

````python
# Example: basic pie chart
values = [5,8,9,10,4,7]
plt.pie(values)
plt.show()
````

But that isn't very useful, is it? Let's look at the same data but with some keyword arguments added to make it more useful.

````python
# Example: better pie chart
values = [5,8,9,10,4,7]
colors = ['r', 'b', 'g', 'y', 'c', 'm']
labels= ['A', 'B', 'C', 'D', 'E', 'F']
explode = [0,0.2,0,0,0,0]

plt.pie(values, colors=colors, labels=labels, explode=explode, autopct='%1.1f%%', counterclock=False, shadow=True)
plt.title('Better pie chart')
plt.show()
````

## Histograms: categorical charts from continuous data
In some cases you have to separate numeric data into classes to make it more useful in your analysis. For example, think back to the New York tax return data. There could be a column in that table for the adjusted gross income (AGI) amount for each taxpayer because it is definitely in the original source data. How would you plot it? A line showing all of the AGI amounts for all NY tax returns? That is much less useful than the categories that actually appear in the data table since _each tax return could theoretically contain a unique AGI amount_. Creating the categories like they did and looking at the frequency of returns in that category is more useful. We use a process called **binning** to create categories (the "bins") and frequency count data from the continuous data that underlies it.

MatPlotLib has histogram functionality built in that will automatically create bins for you based upon a few parameters you input.

````python
# Example: binning and plotting exam data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
# Example: auto binning with random numbers
x = 20 * np.random.randn(10000)

plt.hist(x, 25, range = (-50,50), histtype = 'stepfilled', align = 'mid', color = 'g', label = 'Test data')
plt.legend()
plt.title('Step Filled Histogram')
plt.show()
````

For many datasets that option will work well, but we often have data that falls into particularly sized bins. Think about exam scores. We know the different bin characteristics for each grade. If we let matplotlib create the bins, the results would be very different than what we need. We can create our own categorical data and bins using some of the techniques we have covered recently, then plot the results.

### Creating a histogram with custom bins
We will use some data from two sections of a previous semester's QMTH 2330 Exam 3 scores to create bins and display a histogram 

````python
# Read the data from the csv file into a dataframe
df = pd.read_csv("Stats Exam 3 data.csv", names=['Score'])

# add a calculated percent column 
df['Percent'] = np.round((df['Score']/150)*100,1)
````

````python
# Take a look at the data
df.head()
````

````python
df.tail()
````

The raw data is now in a form to use to determine the letter grades. It could also be used with the `hist()` method of pyplot. With this method we specify the data source (df['Percent']), the number of bins desired (6), histogram type, and the alignment of the bars. The bins are then built to be equal in size.

````python
# What would the histogram look like if we used the hist() method?
plt.hist(df['Percent'],6,histtype='stepfilled', align='mid')
plt.show()
````

````python
# specify the cutoff points for the bins
bins = [0,60,70,80,90,100]
letter_grades = pd.cut(df['Percent'], bins)
letter_grades
````

````python
# show the counts in each category
pd.value_counts(letter_grades)
````

---
What's with the open parenthesis `(` and a closed square bracket `]`? That is the notation used to show which end of the interval is inclusive. We want the _left_ side of the intervals to be inclusive, not the right side. Otherwise, a student getting an 80 on the exam will be counted with the Cs.

---

````python
# corrects which side of the interval is inclusive
letter_grades = pd.cut(df['Percent'], bins, right=False)
print(letter_grades)
print(pd.value_counts(letter_grades))
````

````python
# What happened to the last three values? Adjust the last bin value.
bins = [0,60,70,80,90,101]
# include bin names
grade_names = ['F', 'D', 'C', 'B', 'A']
letter_grades = pd.cut(df['Percent'], bins, labels = grade_names, right=False)
print(letter_grades.tail())
print(pd.value_counts(letter_grades))
````

We can now use the counts as the basis for a histogram of the grades.

````python
data = pd.value_counts(letter_grades)
#data.head()
print(data.index)
print(data.values)
````

````python
# Sort the data into the desired order
data.sort_index(ascending = True, inplace=True)
````

````python
# Plot the histogram 
plt.bar(data.index, data.values, align='center')
plt.title('Exam 3 Grade Distrobution')
plt.show()
````

````python
# Example: a more complete histogram
ax = plt.axes()
ax.set(title = 'Exam 3 Grade Distrobution', ylabel = 'Countt', xlabel = 'Letter Grade')

colors = ['c', 'c', 'c', 'orange', 'khaki']
plt.bar(data.index, data.values, align='center', color = colors, edgecolor = 'k', width = 1)
plt.annotate('Problem', xy=[4,25], xytext = [3,120], arrowprops = dict(facecolor='gray', shrink = 0.05))
plt.show()
````

## Scatterplots: looking for patterns
A scatterplot is great way to get an idea of what your data looks like when you have two-dimensional data. A basic scatterplot is just a graphic representation of each data point on an xy axis. We use them to see if there are any patterns in the data.

### Basic scatterplot

````python
# Example: basic scatterplot
df2 = pd.read_csv('Stats HW and Exams.csv')
df2.head()
````

````python
# Storing the data to two lists for use in the scatterplot makes
#   the code a little cleaner.
x = df2.Homework
y = df2.Exams
````

````python
# Create the basic scatterplot
plt.scatter(x,y)

z = np.polyfit(x,y,1)
p = np.poly1d(z)

plt.plot(x,p(x), 'm--')

plt.show()
````

## Time series

Data gathered over time represents a time series. When we visualize this type of data we often need one of the axes to be time. This example using electricity usage data from London shows usage over time.

Data source: https://data.london.gov.uk/dataset/smartmeter-energy-use-data-in-london-households

Code adapted from https://towardsdatascience.com/playing-with-time-series-data-in-python-959e2485bff8

````python
# Read the data file and see what the structure looks like
dfts = pd.read_csv("UKPN-LCL-smartmeter-sample.csv")
dfts.head()
````

````python
# We want to rename the column containing the usage rates to 'usage' for simplicity
dfts.rename(index=str,columns={'KWH/hh (per half hour) ':'usage'}, inplace = True)

dfts.info()
````

Here are some helpful links on topics used in the following cells.

- Using the loc method of the DataFrame to extract specific pieces of data: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.loc.html

- The `to_numeric` function: http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_numeric.html

- The `numpy.arange` function: https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html

````python
# create date and time columns from DateTime
# DateTime is a string object, not a date
# pandas has a built-in function to handle the conversion
dfts['date'] = pd.to_datetime(dfts['DateTime'], dayfirst = True)

# create a DataFrame to hold just the data we want
data = dfts.loc[:,['usage']]          # gets a slice of all rows and just the usage column
data = data.set_index(dfts.date)          # sets the date as the index

# convert usage data from object to float
data['usage'] = pd.to_numeric(data['usage'], downcast = 'float', errors = 'coerce')
````

````python
# Check the contents of the dataframe
data.head()
#data.tail()
````

````python
# We can get a basic line chart by calling the plot method of the dataframe
data.plot()
````

````python
# Look at weekly data by calling the resample() method
weekly = data.resample('W').sum()
weekly.plot(style = ['--'], figsize = (15,10))
````
